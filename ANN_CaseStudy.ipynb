## Data loading

### Subtask:
Load the "recipes.csv" and "test_recipes.csv" datasets.


```python
import pandas as pd
import os

try:
    df_recipes = pd.read_csv("recipes.csv")
    df_test_recipes = pd.read_csv("test_recipes.csv")
    print("Successfully loaded datasets.")
    display(df_recipes.head())
    display(df_test_recipes.head())
except FileNotFoundError:
    print("Error: One or both CSV files not found in the current working directory.")
    print(f"Current working directory: {os.getcwd()}")
except pd.errors.ParserError:
    print("Error: Could not parse the CSV file. Please check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

## Data preparation

### Subtask:
Combine, clean, and pre-process the recipe datasets.


```python
import pandas as pd
import numpy as np

# Combine DataFrames
combined_df = pd.concat([df_recipes, df_test_recipes], ignore_index=True)

# Handle Missing Values (Initial)
# Fill numeric columns with 0.  More sophisticated imputation could be done later
numeric_cols = ['servings', 'rating']
combined_df[numeric_cols] = combined_df[numeric_cols].fillna(0)

# Drop rows with missing 'ingredients' or 'directions'
combined_df.dropna(subset=['ingredients', 'directions'], inplace=True)

# Data Type Conversion
combined_df['servings'] = pd.to_numeric(combined_df['servings'], errors='coerce')

# Text Data Cleaning (Initial)
# Lowercase and remove special characters from 'ingredients' and 'directions'
for col in ['ingredients', 'directions']:
    combined_df[col] = combined_df[col].astype(str).str.lower()
    combined_df[col] = combined_df[col].str.replace('[^a-zA-Z0-9\s]', '', regex=True)

display(combined_df.head())
```

```python
import re
from sklearn.preprocessing import MinMaxScaler

# Further Text Data Cleaning
def clean_ingredients(ingredients):
    ingredients = re.sub(r'[^\w\s]', '', ingredients) #remove punctuation
    ingredients = re.sub(r'\d+', '', ingredients) #remove digits
    ingredients = ingredients.lower() #lowercase
    return ingredients
combined_df['ingredients'] = combined_df['ingredients'].apply(clean_ingredients)


# Categorical Feature Encoding (Example with cuisine_path)
# Assuming 'cuisine_path' is a categorical feature
combined_df['cuisine_path'] = combined_df['cuisine_path'].astype(str)

# Numerical Feature Scaling (Example with 'servings')
scaler = MinMaxScaler()
combined_df[['servings']] = scaler.fit_transform(combined_df[['servings']])

display(combined_df.head())
```

## Data splitting

### Subtask:
Split the combined recipe dataset into training and testing sets.


```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets (70/30 split)
df_train, df_test = train_test_split(combined_df, test_size=0.3, random_state=42, stratify=combined_df['cuisine_path'])

# Print the shapes of the resulting DataFrames
print("df_train shape:", df_train.shape)
print("df_test shape:", df_test.shape)
```

```python
from collections import Counter

# Count occurrences of each cuisine
cuisine_counts = Counter(combined_df['cuisine_path'])

# Identify cuisines with only one member
cuisines_to_remove = [cuisine for cuisine, count in cuisine_counts.items() if count == 1]

# Remove rows with those cuisines
combined_df_filtered = combined_df[~combined_df['cuisine_path'].isin(cuisines_to_remove)]

# Split the filtered data
df_train, df_test = train_test_split(combined_df_filtered, test_size=0.3, random_state=42, stratify=combined_df_filtered['cuisine_path'])

# Print the shapes of the resulting DataFrames
print("df_train shape:", df_train.shape)
print("df_test shape:", df_test.shape)
```

## Feature engineering

### Subtask:
Extract and engineer relevant features from the training and testing datasets for the subsequent modeling steps.  Focus on creating features suitable for CNNs (image features), NLP models (text features from recipes), and recommendation models.


```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Image Feature Extraction (Placeholder)
df_train['image_features'] = [np.random.rand(128) for _ in range(len(df_train))]
df_test['image_features'] = [np.random.rand(128) for _ in range(len(df_test))]

# 2. Text Feature Engineering
# Ingredient Embeddings
tfidf_ingredients = TfidfVectorizer(max_features=100)  # Adjust max_features as needed
train_ingredients_embeddings = tfidf_ingredients.fit_transform(df_train['ingredients']).toarray()
test_ingredients_embeddings = tfidf_ingredients.transform(df_test['ingredients']).toarray()
df_train['ingredient_embeddings'] = list(train_ingredients_embeddings)
df_test['ingredient_embeddings'] = list(test_ingredients_embeddings)

# Recipe Description Embeddings
tfidf_directions = TfidfVectorizer(max_features=100)  # Adjust max_features as needed
train_directions_embeddings = tfidf_directions.fit_transform(df_train['directions']).toarray()
test_directions_embeddings = tfidf_directions.transform(df_test['directions']).toarray()
df_train['directions_embeddings'] = list(train_directions_embeddings)
df_test['directions_embeddings'] = list(test_directions_embeddings)


# 3. Recommendation Features
df_train['combined_features'] = df_train['servings'] * df_train['rating']
df_test['combined_features'] = df_test['servings'] * df_test['rating']

# 4. Dataframe Consolidation and Verification
print("df_train shape:", df_train.shape)
print("df_test shape:", df_test.shape)
display(df_train.head())
display(df_test.head())
```

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split

# 1. CNN Model (Image Recognition)
X_cnn_train = np.array(list(df_train['image_features']))
y_cnn_train = df_train['rating']
X_cnn_test = np.array(list(df_test['image_features']))
y_cnn_test = df_test['rating']
cnn_model = RandomForestClassifier(random_state=42)
cnn_model.fit(X_cnn_train, y_cnn_train)

cnn_predictions = cnn_model.predict(X_cnn_test)
print("CNN Model Evaluation:")
print("Accuracy:", accuracy_score(y_cnn_test, cnn_predictions))
print("Precision:", precision_score(y_cnn_test, cnn_predictions, average='weighted'))
print("Recall:", recall_score(y_cnn_test, cnn_predictions, average='weighted'))
print("F1-Score:", f1_score(y_cnn_test, cnn_predictions, average='weighted'))

# 2. NLP Model (Recipe Parsing)
X_nlp_train = np.array(df_train['ingredient_embeddings'].tolist())
y_nlp_train = df_train['rating']
X_nlp_test = np.array(df_test['ingredient_embeddings'].tolist())
y_nlp_test = df_test['rating']
nlp_model = LogisticRegression(max_iter=1000, random_state=42)
nlp_model.fit(X_nlp_train, y_nlp_train)
nlp_predictions = nlp_model.predict(X_nlp_test)
print("\nNLP Model Evaluation:")
print("Accuracy:", accuracy_score(y_nlp_test, nlp_predictions))
print("Precision:", precision_score(y_nlp_test, nlp_predictions, average='weighted'))
print("Recall:", recall_score(y_nlp_test, nlp_predictions, average='weighted'))
print("F1-Score:", f1_score(y_nlp_test, nlp_predictions, average='weighted'))


# 3. Recommendation Model (Meal Suggestions)
X_rec_train = df_train[['servings', 'rating']]
y_rec_train = df_train['combined_features']
X_rec_test = df_test[['servings', 'rating']]
y_rec_test = df_test['combined_features']
rec_model = RandomForestClassifier(random_state=42)
rec_model.fit(X_rec_train, y_rec_train)
rec_predictions = rec_model.predict(X_rec_test)
print("\nRecommendation Model Evaluation:")
print("Accuracy:", accuracy_score(y_rec_test, rec_predictions))
print("Precision:", precision_score(y_rec_test, rec_predictions, average='weighted'))
print("Recall:", recall_score(y_rec_test, rec_predictions, average='weighted'))
print("F1-Score:", f1_score(y_rec_test, rec_predictions, average='weighted'))
```

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 1. CNN Model (Image Recognition)
X_cnn_train = np.array(list(df_train['image_features']))
y_cnn_train = df_train['rating']
X_cnn_test = np.array(list(df_test['image_features']))
y_cnn_test = df_test['rating']
cnn_model = RandomForestRegressor(random_state=42)
cnn_model.fit(X_cnn_train, y_cnn_train)
cnn_predictions = cnn_model.predict(X_cnn_test)
print("CNN Model Evaluation:")
print("Mean Squared Error:", mean_squared_error(y_cnn_test, cnn_predictions))
print("R-squared:", r2_score(y_cnn_test, cnn_predictions))

# 2. NLP Model (Recipe Parsing)
X_nlp_train = np.array(df_train['ingredient_embeddings'].tolist())
y_nlp_train = df_train['rating']
X_nlp_test = np.array(df_test['ingredient_embeddings'].tolist())
y_nlp_test = df_test['rating']
nlp_model = LinearRegression()
nlp_model.fit(X_nlp_train, y_nlp_train)
nlp_predictions = nlp_model.predict(X_nlp_test)
print("\nNLP Model Evaluation:")
print("Mean Squared Error:", mean_squared_error(y_nlp_test, nlp_predictions))
print("R-squared:", r2_score(y_nlp_test, nlp_predictions))


# 3. Recommendation Model (Meal Suggestions)
X_rec_train = df_train[['servings', 'rating']]
y_rec_train = df_train['combined_features']
X_rec_test = df_test[['servings', 'rating']]
y_rec_test = df_test['combined_features']
rec_model = RandomForestRegressor(random_state=42)
rec_model.fit(X_rec_train, y_rec_train)
rec_predictions = rec_model.predict(X_rec_test)
print("\nRecommendation Model Evaluation:")
print("Mean Squared Error:", mean_squared_error(y_rec_test, rec_predictions))
print("R-squared:", r2_score(y_rec_test, rec_predictions))
```
